### Task (a)

(i)<br/>
To find the plane tangent to $Y(L, K)$ at point (1000 hours, \$1 million), we need to find the partial derivatives for variables L and K.
$$ \dfrac{\partial{Y(10^3, 10^6)}}{\partial{L}} = 1.5 \dfrac{K^{0.25}}{L^{0.25}} = 1.5*10^{0.75} $$
$$ \dfrac{\partial{Y(10^3, 10^6)}}{\partial{K}} = 0.5 \dfrac{L^{0.75}}{K^{0.75}} = \dfrac{0.5}{10^{2.25}} $$

Then, we plug these values into the equation of a tangent plane:
$$ Y_{tan}(L, K) = Y(L_0, K_0) + \dfrac{\partial{Y(10^3, 10^6)}}{\partial{L}}(L - L_0) + \dfrac{\partial{Y(10^3, 10^6)}}{\partial{K}}(K - K_0) $$
$$  Y_{tan}(L, K) = 10^{6.75} + 1.5*10^{0.75}(L - 10^3) + \dfrac{0.5}{10^{2.25}}(K - 10^6) $$

(ii)<br/>
To find the linear approximation at $Y(10^3 + 1, 10^6 + 1)$, we will plug these values in instead of L and K in the equation above:
$$ Y_{tan}(10^3 + 1, 10^6 + 1) = 2*10^{3.75} + 1.5*10^{0.75}*1 + \dfrac{0.5}{10^{2.25}}*1 ≈ 11255.2644 $$

The change then would be:
$$ Y_{tan}(10^3 + 1, 10^6 + 1) - Y(10^3, 10^6) ≈ 8.4379 $$

(iii)<br/>
The actual function value at point (10^3 + 1, 10^6 + 1) is:
$$ Y(10^3 + 1, 10^6 + 1) = 2 * (10^3 + 1)^{0.75} * (10^6 + 1)^{0.25} ≈ 11246.8265 $$

Then, the actual change is:
$$ Y(10^3 + 1, 10^6 + 1) - Y(10^3, 10^6) = 8.4368 $$

Let's find the relative error:
$$ \dfrac{|Y_{tan}(10^3 + 1, 10^6 + 1) - Y(10^3 + 1, 10^6 + 1)|}{Y(10^3 + 1, 10^6 + 1) - Y(10^3, 10^6)} ≈ 0.0125% $$

This relative error is very small, so we can count our approximation as very decent. It is quite expectable because the changes in 1 hour compared to 1000 hours and \$1 compared to \$1M are very tiny, so that new point is within the boundaries of (1000 h, $1M), where the linear approximation does a very good job.

## A.2
Heavy Metal (***#diffapplication***, ***#difftheory***)

### Task (a)

To find the rate of change of T when moving towards a certain point (directional derivative), we need to first find the vector of movement.

Begin point: (1, 2, 2), end point: (2, 1, 3). Vector of movement: $\vec{v} = < 2 - 1, 1 - 2, 3 - 2 > = < 1, -1, 1 >$

Then, to transform this direction vector into a unit vector, we will divide all its coordinates by its magnitude:
$$ ||\vec{v}|| = |v| = \sqrt{1^2 + (-1)^2 + 1^2} = \sqrt{3} $$
$$ \vec{u} = \dfrac{\vec{v}}{|v|} = < \dfrac{1}{\sqrt{3}}, -\dfrac{1}{\sqrt{3}}, \dfrac{1}{\sqrt{3}} > $$

Since T is inversely proportional to the distance from the center, then we can infer that the formula for the temperature could be:

$$ T = \dfrac{T_0}{1 + \sqrt{x^2 + y^2 + z^2}} $$ where $T_0$ is the temperature at the origin. 1 is added to the denominator to avoid division by zero when the point is at the origin.
$$ \dfrac{T_0}{1 + \sqrt{1^2 + 2^2 + 2^2}} = 120^\circ $$
$$ T_0 = 120 + 120 * 3 = 480^\circ $$

Next step we would like to find all partial derivatives of T to find its gradient:
$$ \dfrac{\partial T}{\partial x} = -\dfrac{480x}{(x^2+y^2+z^2)^{3/2}} $$
$$ \dfrac{\partial T}{\partial y} = -\dfrac{480y}{(x^2+y^2+z^2)^{3/2}} $$
$$ \dfrac{\partial T}{\partial z} = -\dfrac{480z}{(x^2+y^2+z^2)^{3/2}} $$
$$ \nabla T = < -\dfrac{480x}{(x^2+y^2+z^2)^{3/2}}, -\dfrac{480y}{(x^2+y^2+z^2)^{3/2}}, -\dfrac{480z}{(x^2+y^2+z^2)^{3/2}} > $$

Now we can compute the directional derivative by "projecting" the gradient onto the direction unit vector:

$$ D_{\vec{u}} T(1, 2, 2) = \nabla T \circ \vec{u} = -480 * \dfrac{1*\dfrac{1}{\sqrt{3}} + 2*(-\dfrac{1}{\sqrt{3}}) + 2*\dfrac{1}{\sqrt{3}}}{(1 + 4 + 4)^{3/2}} = -\dfrac{480}{\sqrt{3} * 27} = -\dfrac{160}{9 \sqrt{3}} (^\circ)$$

### Task (b)

First, let's find out why the gradient vector is pointing to the origin in this case. As we've discussed before, the gradient vector is:
$$ \nabla T = < -\dfrac{480x}{(x^2+y^2+z^2)^{3/2}}, -\dfrac{480y}{(x^2+y^2+z^2)^{3/2}}, -\dfrac{480z}{(x^2+y^2+z^2)^{3/2}} > $$
$$ \nabla T = \dfrac{480}{(x^2+y^2+z^2)^{3/2}} < -x, -y, -z > $$

Recalling that $\nabla T$ is a gradient from coordinates (x, y, z), we can easily see that this is a parallel vector with the opposite direction, scaled by some positive value. Since $< x, y, z >$ is the vector lying on the line going through (0, 0, 0) and (x, y, z), $k * < -x, -y, -z>$ would be a vector lying on a parallel line. Since we know this this vector passes through (x, y, z) (since we calculate gradient from that point), it will surely pass (0, 0, 0), meaning that it points "towards" the origin of the metal ball.

When we find the directional derivative from any point in the ball, we project the gradient vector on the direction vector. If we project the gradient vector on the vector with the same direction as itself, then the projected length will be maximized.

Gradient itself tells us about how fast the function is changing in each direction. This also gives us information about which axes to "prioritize" because if the function changes the fastest in direction X then we should move along it. In this case, the gradient would point to the center because the most changes would happen when we reduce the sum of squares the most.

Formula-wise, we know that the angle between two vectors can be defined as:
$$ cos \theta = \dfrac{\vec{u} \circ \vec{v}}{|u||v|} $$
$$ |u||v|cos \theta = \vec{u} \circ \vec{v} $$
We can see that the left expression is maximized when $ cos \theta = 1 $ ($ \theta = 0 $). This means that the right expression is maximized when the angle between the two vectors is zero, which is when they are parallel to each other.

### Task (c)

**(1) Using related rates**

If we introduce the variable t which is the time. Then $\dfrac{df}{dt}$ would represent the rate of change of the conversion w.r.t. time.

$$ f(x, y) = \dfrac{y}{x} $$
$$ \dfrac{d f(x, y)}{d t} = \dfrac{\frac{d y}{d t} x - \frac{d x}{d t} y}{x^2}, \dfrac{d y}{d t} = 0 $$
$$ \dfrac{d f(x, y)}{d t} = -\dfrac{d x}{d t} * \dfrac{y}{x^2} = 0.05x \dfrac{y}{x^2} = 0.05\dfrac{y}{x}  $$'

This means that when the conversion rate decreases, we will get more Dollars for the Euros we have, specifically, we will get 5% more.

**(2) Using partial derivatives**

Equation of the plane, tangent at point $(x_0, y_0)$:
$$ f_{tan}(x, y) = f(x_0, y_0) + \dfrac{\partial f(x_0, y_0)}{\partial x}(x - x_0) + \dfrac{\partial f(x_0, y_0)}{\partial y}(y - y_0) $$
$$ f_{tan}(x, y) = \dfrac{y_0}{x_0} - \dfrac{y_0}{x_0^2}(x - x_0) + \dfrac{1}{x_0}(y - y_0) $$

Since we get a 5% decrease, we will put in $x = 0.95x_0$:
$$ f_{tan}(0.95x_0, y_0) = \dfrac{y_0}{0.95x_0} + \dfrac{y_0}{x_0^2}0.05x_0 + 0 = \dfrac{419}{380} \dfrac{y_0}{x_0} $$
$$ \Delta f = f_{tan}(0.95x_0, y_0) - f(x_0, y_0) ≈ 0.10 \dfrac{y_0}{x_0} $$

This is a linear approximation of how the function changes when x decreases by 5% (not so precise, since it should increase by 5%, not 10%).

**(3) Using partial derivatives (2)**

$$ \dfrac{\partial f(x_0, y_0)}{\partial x} = -\dfrac{y_0}{x_0^2} $$
$$ \partial f(x_0, y_0) = -\partial x \dfrac{y_0}{x_0^2} $$
$$ \partial f(x_0, y_0) = 0.05x \dfrac{y_0}{x_0^2} = 0.05 \dfrac{y_0}{x_0} $$

Same result as in the first one, but now using the definition of a derivative and differential terms.

***Figure 2. Tangent plane intercepting the conversion function at point (0.81, 200). Using this plane, we then make a linear approximation for the situation when conversion rate is 0.77 $\dfrac{Euro}{Dollar}$, and the amount exchanged is 150 $Euro$.***

To estimate the combined effects of the discount and the new conversion factor, let's place these values in the tangent plane equation. In this case, we initially knew that conversion rate is 0.81 Euro and product price is 200 Euro and would like to estimate what would be the change if the new conversion rate is 0.77 and product price is 150.

$$ f_{tan}(0.77, 150) = \dfrac{200}{0.81} - \dfrac{200}{0.81^2}(0.77 - 0.81) + \dfrac{1}{0.81}(150 - 200) ≈ 197.3785 $$
Then, the combined effects is the difference between the old price and the new price:
$$ \Delta_{est} = |f(0.81, 200) - f_{tan}(0.77, 150)| ≈ 49.5351 $$

Now, the real difference is:
$$ \Delta_{actual} = |f(0.81, 200) - f(0.77, 150)| ≈ 52.1084 $$

The relative error is:
$$ \epsilon = \dfrac{|\Delta_{est} - \Delta_{actual}|}{\Delta_{actual}} ≈ 4.94\% $$

Quite a good estimate, off by only 5%! Most likely what happened here is that the more we go off from the original ratio ($\dfrac{0.81}{200}$), the less accurate the prediction will be. For example, if we take a look at $(0.6, 220)$ in the graph, the distance is big, because we have decreased the conversion rate, increasing the function, and we increased the converted sum, increasing the function again. Thus, we have diverged from the original ratio, for which we know the exact answer, so then the estimate is worse.
