```python
var('sd_x,sd_y,ro')
sd_x = 1
sd_y = 1
ro = 0

z(x,y) = x^2/sd_x^2 - ro*x*y/(sd_x*sd_y) + y^2/sd_y^2
P(x,y) = 1/(2*pi*sd_x*sd_y*sqrt(1-ro^2)) * e^(-z(x,y) / (2 * (1-ro^2)))

contour_plot(P(x,y), (x,-3,3), (y,-3,3), frame=True, axes_labels=['x','y'], axes=False, labels=True, title='Contour plot of probability density function (PDF) of two variables.', title_pos=(0.5,-0.15))
```

```python
print(integrate(P, x, -oo, oo)) # = 1 / sqrt(2*pi)
print(integrate(sqrt(2*pi) * P, x, -oo, oo)) # = 1
```

It is almost a PDF of one variable (when $\sigma_y = 1$ and $ \rho = 0 $) because when y = 0, all the terms related to it cancel to zero. Then, we are left with $ P(x, 0) = \dfrac{1}{2 \pi \sigma_x} * exp [-\frac{x^2}{2\sigma_x^2}] $. The area that we get when we integrate this 2D distribution is $ \dfrac{1}{\sqrt{2 \pi}} $ (which is what the above thing is equal to, just replace y with 0). Therefore, we would need to mulptiply the PDF function by $ \sqrt{2 \pi} $ to get the sum of all probabilities = 1. Then, this would be a PDF function of variable x.

```python
var('sd_x,sd_y,ro')
sd_x = 1
sd_y = 1
ro = 0

z(x,y) = x^2/sd_x^2 - ro*x*y/(sd_x*sd_y) + y^2/sd_y^2
P(x,y) = 1/(2*pi*sd_x*sd_y*sqrt(1-ro^2)) * e^(-z(x,y) / (2 * (1-ro^2)))

plot(P(x,1), (x,-5,5), axes_labels=['x', 'z'], axes=True)
```

```python
print(integrate(P, x, -oo, oo)) # = 1 / (e^0.5 * sqrt(2*pi))
print(integrate(e**.5 * sqrt(2*pi) * P, x, -oo, oo)) # = 1
```

When y = 1, $ \sigma_y = 1, \rho = 0 ⇒ P(x, 1) = \dfrac{1}{2 \pi \sigma_x} * exp [-\frac{x^2 + 1}{2\sigma_x^2}] = \dfrac{1}{2 \pi \sigma_x e^{\frac{1}{2\sigma_x}}} * exp [-\frac{x^2}{2\sigma_x^2}] $ - this is proportional to the first one, just squished in the z-axis. Can still work as a PDF for x if we multiply the PDF function by $ e^{\frac{1}{2}} \sqrt{2 \pi} $ to equalize it with the single-variable PDF function above.

```python
var('c, L_0')
c = 3 * 10^8 # m/s
L_0 = 100 # spaceship is 100m long at rest

L(v) = L_0 * sqrt(1 - v^2/c^2)

plot(L(v), (v,0,3*10^8), frame=True, axes_labels=['Speed (m/s)', 'Observable length (m)'], axes=False, title='Observable length of a moving spaceship based on special relativity principle.', title_pos=(0.5,-0.15))
```

### Task (b)

As $v$ increases, $L$ decreases subquadratically ⇒ the higher is the speed (closer is to the speed of light), the higher is the rate, at which L drops. This makes sense because $ L'(v) = -\dfrac{L_o v}{c^2 \sqrt{1 - \frac{v^2}{c^2}}} $ ⇒ as $ v $ increases, the numerator will increase, and the denominator will decrease. We can also notice that observable length $ L = 0 $ at $ v = 3*10^8 $. 

### Task (c)

We can observe that in the graph, the speed tends to zero when the speed of the ship is near the speed of light. This can also be observed in the formula: this function's domain $ D = [0; 3*10^8] $ ⇒ it is continuous at the point $3*10^8$ ⇒ $\lim_{v\to c-} L_0*\sqrt{1-\dfrac{v^2}{c^2}}= L(c) = 0$

```python
lim(L(v), v=c, dir='-') # v -> c-
```

Ceiling function is useful when we would like to provide an upper bound estimation. For example, each sofa can fit k people, and we have n people. We would like to have everyone seated, even if one sofa might be half-empty. Therefore, we would do tmp = n / k, num_of_sofas = ceil(tmp)<br/>

Lastly, signum function is used in perceptron (or more generally, as an activation function for artificial neurons).
```python
# Let's imagine we have such a function:
plot(sgn(x)*(ceil(x) - floor(x) - 1), (x, -5, 5))
```

```python
# We can observe for all integer x, ceil(x) = floor(x)
# Therefore, function value = -1*sgn(x) != 0 as other points
# Thus, this function is not continuous
# Now, we would like to somehow redefine the function when x in Z
# To do so, let's figure out how to determine if x is an integer

var = 6.5 # some number to work with
is_int = var % 2 # integer would give 0 or 1 while real would give smth in b/w
is_int = abs(is_int - 0.5) # 0, 1 -> 0.5, real -> != 0.5

# Now we would want a function that can transform 0.5 into 1 and != 0.5 into 0 => PDF! Almost...
f = 1/(sqrt(2*pi)) * e^(-is_int^2) # peak is at 1 / sqrt(2*pi), we want 1
f = e^(-is_int^2) # yay, now let's center it at 0.5
f = e^(-(is_int - 0.5)^2) # now how do we transform as promised??
f = floor(e^(-(is_int - 0.5)^2)) # wow, x < 0.5 -> f = 0, x = 0.5 -> f = 1

# So what do we want to do with this?
# Well, just add it to make the original function 0 instead of -1 when x is int
plot(sgn(x)*(ceil(x) - floor(x) - 1 + 1 * (floor(e^(-(abs(x - floor(x/2)*2 - 0.5) - 0.5)^2)))), (x, -5, 5))
```

### Task (b)

(1) $ lim_{(x, y) \to (\infty, \infty)} 5 + \dfrac{sin4x*sin3x}{1+x^2+y^2} = lim_{(x, y) \to (\infty, \infty)} 5 + \dfrac{1}{1+x^2+y^2} $ (since $ sin4x*sin3y \le 1 * 1 = 1 $) = $ lim_{(x, y) \to (\infty, \infty)} 5 + 0  = 5 $<br/>

(2) $ lim_{(x, y) \to (0, \infty)} 5 + \dfrac{sin4x*sin3x}{1+x^2+y^2} = lim_{(x, y) \to (0, \infty)} 5 + \dfrac{1}{1+x^2+y^2} $ (since $ sin4x*sin3y \le 1 * 1 = 1 $) = $ lim_{(x, y) \to (0, \infty)} 5 + 0  = 5 $<br/>

(3) $ lim_{(x, y) \to (\infty, 0)} 5 + \dfrac{sin4x*sin3x}{1+x^2+y^2} = lim_{(x, y) \to (\infty, 0)} 5 + \dfrac{1}{1+x^2+y^2} $ (since $ sin4x*sin3y \le 1 * 1 = 1 $) = $ lim_{(x, y) \to (\infty, 0)} 5 + 0  = 5 $<br/>

It's worth noticing that no matter what configuration of infinity we have for (x, y), as long as one of them approaches infinity, the denominator becomes infinity, making the fraction equal to zero.<br/>
