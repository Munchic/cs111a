### Algorithm to finding a secant line
1. Determine 2 points $(x_0, y_0)$ and $(x_1, y_1)$ on the function that the secant line should pass through<br/>
2. Find the slope of the line that passes through these two points, $m = \dfrac{y_1 - y_0}{x_1 - x_0}$</br>
3. Let the secant line equation pass through one of the points (e.g. $(x_0, y_0)$) and write down equation: $ y_{sec} - y_0 = m * (x - x_0) $
4. Solve $ y_{sec} $ in terms of $x$: $ y_{sec} = m*x - m*x_0 + y_0 $ by placing values of $ (x_0, y_0) $ in.

### Algorithm to finding a tangent line
1. Determine a point $(x_0, y_0)$ on the function that the tangent line should pass through<br/>
2. Find the slope (rate of change, or, the derivative) of the line that passes through this point, $k = f'(x_0)$</br>
3. Let the tangent line equation pass through the point $(x_0, y_0)$ and write down equation: $ y_{sec} - y_0 = k * (x - x_0) $
4. Solve $ y_{sec} $ in terms of $x$: $ y_{sec} = k*x - k*x_0 + y_0 $ by placing values of $ (x_0, y_0) $ in.
